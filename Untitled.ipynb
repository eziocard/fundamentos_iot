```python
import pandas as pd




df = pd.read_csv('datos_sensores.csv', na_values=['NA', 'null', 'N/A', ''])
humedad_aire = df['humedad_aire']
temperatura = df['temperatura']
humedad_tierra = df['humedad_tierra']

humedad_aire = df['humedad_aire'].dropna()
temperatura = df['temperatura'].dropna()
humedad_tierra = df['humedad_tierra'].dropna()

humedad_aire = humedad_aire.reset_index(drop=True)
temperatura = temperatura.reset_index(drop=True)
humedad_tierra = humedad_tierra.reset_index(drop=True)

new_data = {
    'humedad_aire': humedad_aire,
    'temperatura': temperatura,
    'humedad_tierra': humedad_tierra
}

ndf= pd.DataFrame(new_data)
print(ndf.to_string())
ndf.to_csv('datos_procesados.csv', index=False)
```
